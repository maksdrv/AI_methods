```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
```

```python
#  4
x_objects = np.random.uniform(0, 1, size=500)
y_objects = 0.1 * np.log(x_objects) + np.random.normal(scale=0.7, size=x_objects.shape)
plt.scatter(x_objects, y_objects, alpha = 0.3)
```

```python
C=np.cov(x_objects,y_objects)
a=C[0,1]/C[0,0]
b=np.mean(y_objects)-a*np.mean(x_objects)
xfit = np.linspace(0, 1, 2)
yfit = a*xfit + b
print(f"a={a}")
print(f"b={b}")
plt.figure()
plt.scatter(x_objects, y_objects, alpha = 0.3)
plt.plot(xfit, yfit, color = "r")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(x_objects[:, np.newaxis], y_objects)
xfit = np.linspace(0, 1, 2)
yfit = model.predict(xfit[:, np.newaxis])
plt.figure()
plt.scatter(x_objects, y_objects, alpha = 0.3)
plt.plot(xfit, yfit, color = "r")
print(f"a={model.coef_[0]}")
print(f"b={model.intercept_}")
```

```python
def ADALINE(w, b, X, Y, alpha):
  for k in range(25):
    for i in range(len(X)):
      y = X[i] * w + b
      w += 2 * alpha * (Y[i] - y) * X[i]
      b += 2* alpha * (Y[i] - y)
  return w, b

a, b = 0, 0
alpha = 0.001
res_a, res_b = ADALINE(a, b, x_objects, y_objects, alpha)
print(f"a = {res_a}")
print(f"b = {res_b}")
x = np.linspace(0, 1, 2)
y = res_a*x+res_b
plt.scatter(x_objects, y_objects, alpha = 0.3)
plt.plot(x, y, color = "r")
```
